```python
class FLNPF(torch.nn.Module):
    def __init__(self,x,y):
        super(FLNPF, self).__init__()
        self.gating_layers = []
        self.value_layers = []
        # self.n=x
        # self.m=y

        self.gating_layers.append(nn.Linear(x, y, 'gating_layer'))
        self.value_layers.append(nn.Linear(x, y))
        for i in range(4):
            self.gating_layers.append(nn.Linear(y, y, 'gating_layer'))
            self.value_layers.append(nn.Linear(y, y))

        self.gating_output_layer = nn.Linear(y, 10)
        self.value_output_layer = nn.Linear(y, 10)

    def forward(self, x, y,running_loss,criterion,optimizer, train_gating_network=False):
        # running_loss=0.0
        if train_gating_network:
            for layer in self.gating_layers:
                x = layer(x)
                x = torch.relu(x)
        else:
            x_g, x_v = x, x
            for layer_g, layer_v in zip(self.gating_layers, self.value_layers):
                x_g = torch.relu(layer_g(x_g))
                x_v = layer_v(x_v)
                # with torch.no_grad():
                gate = torch.sign(x_g)
                x_v = x_v * gate.detach()

        
        if train_gating_network:
          outputs = self.gating_output_layer(x)
        else:
          outputs = self.gating_output_layer(x_v)
        outputs=torch.nn.functional.softmax(outputs,dim=1)
        loss = criterion(outputs, y)
        
        # for i in range(len(self.gating_layers)-1,-1,-1):
        # optimizer = torch.optim.Adam(self.gating_layers[i].parameters(), lr=1e-4)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        print()
        return running_loss
        # print(running_loss)

input_size = 784
hidden_sizes = 128
# output_size = 10
epochs=2
model1=FLNPF(input_size,hidden_sizes)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model1.parameters(), lr=1e-4)
for e in range(epochs):
    running_loss=0.0
    for images, labels in trainloader:
        images = images.view(images.shape[0], -1)
        running_loss+=model1.forward(images,labels,running_loss,criterion,optimizer,True)
    print("Epoch {} - Training loss: {}".format(e, running_loss/len(trainloader)))
```
